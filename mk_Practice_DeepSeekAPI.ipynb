```python
import requests

# Configuration
API_KEY = "sk-bd8d7f42788e46068fd2da4d4826f7a8"
API_URL = "https://api.deepseek.com/v1/chat/completions"  # Hypothetical endpoint
MODEL = "deepseek-chat"  # Adjust based on available models

def query_deepseek_api(prompt: str, temperature: float = 0.7) -> str:
    # """
    # Sends a prompt to the DeepSeek API and returns the assistant's reply.
    
    # Args:
    #     prompt (str): User's input/question.
    #     temperature (float): Controls randomness (0-2). Lower = more deterministic.
    
    # Returns:
    #     str: Generated response from DeepSeek.
    # """
    headers = {
        "Authorization": f"Bearer {API_KEY}",
        "Content-Type": "application/json"
    }
    
    payload = {
        "model": MODEL,
        "messages": [{"role": "user", "content": prompt}],
        "temperature": temperature,
    }
    
    try:
        response = requests.post(API_URL, headers=headers, json=payload)
        response.raise_for_status()  # Raise HTTP errors if any
        
        # Extract and return the assistant's reply
        return response.json()["choices"][0]["message"]["content"]
    
    except requests.exceptions.RequestException as e:
        return f"API Error: {e}"

# Example Usage
if __name__ == "__main__":
    user_prompt = "Explain quantum computing in simple terms."
    print("Sending query to DeepSeek API...\n")
    
    reply = query_deepseek_api(user_prompt)
    print("DeepSeek's Response:")
    print("-" * 50)
    print(reply)
    print("-" * 50)
```

```python
import requests
from rdflib import Graph, URIRef, RDF, RDFS, OWL

# --- Step 1: Parse the Turtle (.ttl) Ontology File ---
def parse_ontology(file_path):
    g = Graph()
    g.parse(file_path, format="turtle")
    
    classes = set()
    object_properties = set()
    data_properties = set()
    
    # Extract Classes (owl:Class)
    for s in g.subjects(RDF.type, OWL.Class):
        classes.add(str(s))
    
    # Extract Object Properties (owl:ObjectProperty)
    for s in g.subjects(RDF.type, OWL.ObjectProperty):
        object_properties.add(str(s))
    
    # Extract Data Properties (owl:DatatypeProperty)
    for s in g.subjects(RDF.type, OWL.DatatypeProperty):
        data_properties.add(str(s))
    
    return {
        "classes": sorted(classes),
        "object_properties": sorted(object_properties),
        "data_properties": sorted(data_properties),
    }

# --- Step 2: Query DeepSeek API for Analysis ---
def query_deepseek_api(prompt: str, api_key: str) -> str:
    API_URL = "https://api.deepseek.com/v1/chat/completions"  # Hypothetical endpoint
    
    headers = {
        "Authorization": f"Bearer {api_key}",
        "Content-Type": "application/json"
    }
    
    payload = {
        "model": "deepseek-chat",
        "messages": [{"role": "user", "content": prompt}],
        "temperature": 0.3,  # Lower for structured output
    }
    
    try:
        response = requests.post(API_URL, headers=headers, json=payload)
        response.raise_for_status()
        return response.json()["choices"][0]["message"]["content"]
    except Exception as e:
        return f"API Error: {e}"

# --- Step 3: Main Workflow ---
if __name__ == "__main__":
    #  DeepSeek API key
    DEEPSEEK_API_KEY = "sk-bd8d7f42788e46068fd2da4d4826f7a8r_api_key_here"
    ONTOLOGY_FILE = "my_ontology.ttl"  # Path to your Turtle file
    
    # Parse the ontology
    ontology_data = parse_ontology(ONTOLOGY_FILE)
    
    # Display extracted data
    print("=== Ontology Summary ===")
    print(f"Classes: {len(ontology_data['classes'])}")
    print(f"Object Properties: {len(ontology_data['object_properties'])}")
    print(f"Data Properties: {len(ontology_data['data_properties'])}\n")
    
    # Generate a prompt for DeepSeek
    prompt = f"""
    Analyze this ontology and provide a summary:
    - Classes: {ontology_data['classes']}
    - Object Properties: {ontology_data['object_properties']}
    - Data Properties: {ontology_data['data_properties']}
    
    Questions:
    1. What is the domain of this ontology likely to be?
    2. Are there any notable patterns in property names?
    3. Suggest improvements for better ontology design.
    """
    
    # Query DeepSeek API
    print("Querying DeepSeek API for analysis...\n")
    analysis = query_deepseek_api(prompt, DEEPSEEK_API_KEY)
    
    print("=== DeepSeek Analysis ===")
    print(analysis)
```
